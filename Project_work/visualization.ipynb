### **3. Visualization with Matplotlib**

#### **Basic Plot Types**

##### **a. Line Plot**

- **Use Case**: Track changes over time or ordered data.

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
fig, ax = plt.subplots()
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]
ax.plot(x, y, marker='o', label="Example Line")
ax.set_title("This is a Line Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.legend()
plt.show()
````

##### **b. Scatter Plot**

- **Use Case**: Show relationship between two variables.

````python
fig, ax = plt.subplots()
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]
ax.scatter(x, y, color='blue')
ax.set_title("Scatter Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
plt.show()
````

##### **c. Bar Plot**

- **Use Case**: Compare quantities across categories.

````python
fig, ax = plt.subplots()
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 18]
ax.bar(categories, values, color='purple')
ax.set_title("Bar Plot")
ax.set_xlabel("Categories")
ax.set_ylabel("Values")
plt.show()
````

##### **d. Histogram**

- **Use Case**: Show distribution of a variable.

````python
fig, ax = plt.subplots()
data = [5, 15, 10, 20, 10, 25, 30, 15, 20, 25]
ax.hist(data, bins=5, color='green')
ax.set_title("Histogram")
ax.set_xlabel("Value")
ax.set_ylabel("Frequency")
plt.show()
````

##### **e. Box Plot**

- **Use Case**: Show summary statistics of a variable.

````python

fig, ax = plt.subplots()
data = [[5, 15, 10, 20], [10, 25, 30, 15], [20, 25, 5, 10]]
ax.boxplot(data, labels=['A', 'B', 'C'])
ax.set_title("Box Plot")
plt.show()
````

### **4. Visualization with Seaborn**

**Standard Format**: We'll use `sns.set_theme()` to establish a consistent theme for all plots.

#### **Basic Plot Types**

##### **a. Line Plot**

- **Use Case**: Track changes in data over time.

  ```python
  sns.set_theme(style="darkgrid")
  data = pd.DataFrame({
      'x': [1, 2, 3, 4, 5],
      'y': [10, 15, 13, 17, 20]
  })
  sns.lineplot(data=data, x='x', y='y').set(title="Seaborn Line Plot")
  ```



##### **b. Scatter Plot**

- **Use Case**: Explore relationship between two variables.

  ```python
  sns.set_theme(style="white")
  data = pd.DataFrame({
      'total_bill': [5, 7, 8, 7, 2, 17, 2, 9, 4, 11],
      'tip': [99, 86, 87, 88, 100, 86, 103, 87, 94, 78],
      'day': ['Thur', 'Fri', 'Sat', 'Sun', 'Fri', 'Thur', 'Sun', 'Sat', 'Sun', 'Thur']
  })
  sns.scatterplot(data=data, x="total_bill", y="tip", hue="day").set(title="Seaborn Scatter Plot")
  ```



##### **c. Bar Plot**

- **Use Case**: Compare categorical data.

  ```python
  sns.set_theme(style="whitegrid")
  data = pd.DataFrame({
      'day': ['Mon', 'Tue', 'Wed', 'Thu'],
      'total_bill': [10, 15, 10, 20]
  })
  sns.barplot(data=data, x="day", y="total_bill", ci=None).set(title="Seaborn Bar Plot")
  ```



##### **d. Histogram**

- **Use Case**: Display distribution of a single variable.

  ```python
  sns.set_theme(style="dark")
  data = pd.DataFrame({'total_bill': [5, 15, 10, 20, 10, 25, 30, 15, 20, 25]})
  sns.histplot(data=data, x="total_bill", kde=True).set(title="Seaborn Histogram")
  ```



##### **e. Box Plot**

- **Use Case**: Summary of data distribution, outliers, and quartiles.

  ```python
  sns.set_theme(style="whitegrid")
  data = pd.DataFrame({
      'day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
      'total_bill': [5, 15, 10, 20, 25, 10, 25, 30, 15, 20]
  })
  sns.boxplot(data=data, x="day", y="total_bill").set(title="Seaborn Box Plot")
  ```

````python
import matplotlib.pyplot as plt
import numpy as np

# Set the style and figure size for better visibility
# plt.style.use('seaborn')
plt.rcParams['figure.figsize'] = [12, 6]

# 1. Age Distribution with Mean Line
plt.figure()
plt.hist(data['Age'].dropna(), bins=30, color='skyblue', alpha=0.7, edgecolor='black')
plt.axvline(data['Age'].mean(), color='red', linestyle='--', label=f'Mean Age: {data["Age"].mean():.1f}')
plt.title('Age Distribution of Olympic Athletes')
plt.xlabel('Age')
plt.ylabel('Number of Athletes')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

````python

# 2. Height vs Weight Scatter Plot by Gender
plt.figure()
colors = {'M': 'blue', 'F': 'red'}
for gender in ['M', 'F']:
    mask = data['Sex'] == gender
    plt.scatter(data[mask]['Height'], data[mask]['Weight'], 
               c=colors[gender], alpha=0.5, label=gender)
plt.title('Height vs Weight by Gender')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

````

````python
# 3. Medal Distribution
medal_counts = data['Medal'].value_counts()
plt.figure()
plt.bar(medal_counts.index, medal_counts.values, color=['gold', 'silver', 'darkgoldenrod'])
plt.title('Distribution of Medal Types')
plt.xlabel('Medal Type')
plt.ylabel('Number of Medals')
plt.grid(True, alpha=0.3)
plt.show()

````
